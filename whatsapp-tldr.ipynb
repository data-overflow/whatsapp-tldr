```python
from transformers import pipeline
summarizer = pipeline("summarization", model="philschmid/bart-large-cnn-samsum", device=0)
```

```python
#upload the txt file without media exported using whatsapp
from google.colab import files
uploaded = files.upload()
filename = list(uploaded.keys())[0]
```

```python
import re

def strip_timestamp(chat_file):
    with open(chat_file, 'r', encoding='utf8') as file:
        lines = file.readlines()
    out = []
    for line in lines:
        pattern = r'\d{1,2}/\d{1,2}/\d{2}, \d{1,2}:\d{2} [ap]m - '
        match = re.search(pattern, line)
        if match:
            stripped_line = line[match.end():]
            out.append(stripped_line)
        else:
            out.append(line)
    return out

messages = strip_timestamp(filename)
max_chunk = 1000
current_chunk = 0
chunks = []
chunk_size = []
i = 0
for message in messages:
    i += 1
    if len(chunks) == current_chunk + 1:
        if len(chunks[current_chunk]) + len(message) <= max_chunk:
            chunks[current_chunk] += message
        else:
            current_chunk += 1
            chunks.append(message)
            chunk_size.append(i)
    else:
        chunks.append(message)
        chunk_size.append(i)


def summarize_chunks(chunks, lastn=5, verbose=False):
    n = lastn
    out = []
    while n > 0:
        print("Summarizing chunk (", lastn - n + 1, "/", lastn, ")")
        res = summarizer(chunks[-n], max_length=128, min_length=8, batch_size=8)
        out.append(res[-1]['summary_text']+"\n")
        if verbose:
          print(res[-1]['summary_text'], end="\n\n")
        n -= 1
    print("Saving output to disk...")
    with open('output.txt', 'w', encoding='utf8') as file:
        file.writelines(out)
    print("All tasks done")
    return out


def summarize_lastn(num_messages):
    global chunks, chunk_size
    total_messages = chunk_size[-1]
    if num_messages >= total_messages:
        return len(chunks)
    i = -1
    while (total_messages - chunk_size[i]) < num_messages:
        i -= 1
    result = summarize_chunks(chunks, abs(i))
```

```python
# Enter the last number of unread messages
summarize_lastn(5000)
```

```python
# Print the result
with open("output.txt", 'r') as file:
  print(file.read())
```
